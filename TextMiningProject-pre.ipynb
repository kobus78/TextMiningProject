# Pre-processing (part 1)

```python
PATH = 'input/'
MAIN = '1965-1212_Communion_ENG_12-1201-B123E1R-x'
filename = f'{PATH}{MAIN}.txt'
```

```python
filename
```

```python
#get ENG source file
with open(filename, 'r', encoding='utf-8') as f_eng:
    eng_lines = f_eng.readlines()
eng_title = eng_lines.pop(0) #remove title line
print(eng_lines[0:3])
```

```python
#prepares a single text string in out_lines by joining all sentences
out_lines = []
cur_rsub = ''
sentences = []
for i in range(len(eng_lines)):
    eng_line = eng_lines[i].strip(); print(eng_line)
    if not eng_line: #This is a blank line
        out_lines.append(' '.join(sentences) + ' ')
        cur_rsub = ''
        sentences = []
    else:
        descriptor,text = eng_line.split(" ", 1)
        rsub,sen,typ = descriptor.strip().split('.')
        sentences.append(text)
        cur_rsub = rsub
        #print('sentences:', sentences)
```

```python
#inspect the output
out_lines
```

```python
#write the output '-top' file to the output folder
path,proper = filename.split("/")
sans_ext,ext = proper.split(".")
with open('output/' + sans_ext + '-top.txt', 'w', encoding='utf-8') as f_out:
    f_out.writelines(out_lines)
    f_out.writelines('\n')
```
